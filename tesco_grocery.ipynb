```python
# finding purchasing pattern of food categories for different age groups
```

```python
# Step 1: Create demographic segments based on the proportion of young population
# Calculate the median percentage of young population (age_0_17) to define high and low youth proportion areas
year_borough['perc_young'] = year_borough['age_0_17'] / year_borough['population']
median_perc_young = year_borough['perc_young'].median()

# Define high and low youth proportion areas
year_borough['youth_segment'] = year_borough['perc_young'].apply(lambda x: 'High Youth' if x >= median_perc_young else 'Low Youth')

# Step 2: Calculate average purchase of food categories for each segment
# Define food category columns (example categories)
food_categories = [col for col in year_borough.columns if col.startswith('f_') and col.endswith('_weight')]

# Group data by youth segment and calculate the mean for each food category
food_preferences_by_youth = year_borough.groupby('youth_segment')[food_categories].mean()

food_preferences_by_youth.transpose()  # Transpose for better readability in displaying results
```

## Insight 2

How does the food choices and nutrients (carbs, sugar, fat, saturated fat, protein, fibre) correlate with obesity-related hospitalizations?

```python
# Extract basic nutrients and hospitalization data
nutrient_columns = ['carb','fat', 'sugar', 'protein', 'fibre', 'saturate','prevalence_hospitalizations', 'prevalence_bariatric']

# Create a subset of data for analysis
nutrient_data = year_borough[nutrient_columns]

# Checking for missing values
missing_data = nutrient_data.isnull().sum()

# Displaying the first few rows of our subset and the missing values
nutrient_data.head(), missing_data
```

```python
# Drop rows with any missing values
clean_data = nutrient_data.dropna()

# Calculate Spearman correlation
spearman_correlation = clean_data.corr(method='spearman')

# Display the correlation matrix
spearman_correlation
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the visual style
sns.set(style="white")

# Create a heatmap
plt.figure(figsize=(10, 8))
heatmap = sns.heatmap(spearman_correlation, annot=True, cmap='coolwarm', fmt=".2f", linewidths=1)
heatmap.set_title('Spearman Correlation Heatmap between Nutrients and Hospitalizations')
plt.show()
```

```python
# Creating scatter plots for each nutrient against hospitalization rates
fig, axes = plt.subplots(2, 2, figsize=(14, 12))
fig.suptitle('Scatter Plots of Nutrients vs. Obesity-Related Hospitalization Rates')

# Mapping nutrients to plots
nutrients = ['fat', 'sugar', 'protein', 'fibre']
hospitalization_types = ['prevalence_hospitalizations', 'prevalence_bariatric']
colors = ['blue', 'green', 'red', 'purple']

for i, nutrient in enumerate(nutrients):
    ax = axes[i//2, i%2]
    for j, hosp_type in enumerate(hospitalization_types):
        sns.scatterplot(data=clean_data, x=nutrient, y=hosp_type, ax=ax, color=colors[j], label=hosp_type)
    ax.set_title(f'Scatter Plot: {nutrient} vs. Hospitalization Rates')
    ax.set_xlabel(f'{nutrient} (grams)')
    ax.set_ylabel('Hospitalization Rate (%)')
    ax.legend()

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

* Fat and Sugar: Both show negative correlations with obesity-related hospitalizations (prevalence_hospitalizations and prevalence_bariatric). This suggests that higher intakes of fat and sugar are associated with lower hospitalization rates in this dataset, which is counterintuitive and may be caused due to the insufficiency and incompleteness of the diet data.
* Protein: Displays a positive correlation with hospitalization rates, indicating that higher protein intake correlates with higher obesity-related hospitalizations. This finding could suggest dietary patterns or other lifestyle factors not captured solely by protein intake.
* Fibre: Has a negative correlation with hospitalization rates, which aligns with general health advice that higher fiber intake is beneficial and might reduce obesity-related health issues.


The heatmap clearly displays the Spearman correlation coefficients, emphasizing the relationships between nutrient intake and obesity-related hospitalizations. As noted:

- Red shades indicate positive correlations, particularly visible between protein and hospitalization rates.
- Blue shades show negative correlations, such as those between fat, sugar, and hospitalizations.

The scatter plots provide a visual exploration of how each nutrient relates to obesity-related hospitalization rates. Here are some observations:

- Fat and Sugar: The scatter plots for fat and sugar against hospitalization rates demonstrate negative trends, where higher values of fat and sugar intake are associated with lower rates of hospitalizations. This is unexpected based on typical nutritional advice and could suggest that other factors, such as overall diet quality or socioeconomic factors, may be influencing these results.
- Protein: There is a visible trend where higher protein intake correlates with higher rates of hospitalizations. This might indicate dietary patterns where high protein intake coexists with other lifestyle or health factors contributing to obesity.
- Fibre: The relationship between fibre intake and hospitalization rates does not show a clear trend, but there is a slight tendency for higher fibre intake to associate with lower hospitalization rates, which is consistent with common dietary recommendations.

These findings suggest complex interactions between diet and health outcomes, possibly influenced by a variety of other factors.

```python
# Reshape the Population data to prepare for a line chart with all regions
population_long_chart = pd.melt(total_res_pop, id_vars=['Region name'], value_vars=[str(year) for year in range(1997, 2022)],
                                var_name='Year', value_name='Population')

# Convert 'Year' to integer
population_long_chart['Year'] = pd.to_numeric(population_long_chart['Year'])

# Reshape the GDHI data to prepare for a line chart with all regions
gdhi_long_chart = pd.melt(total_GDHI, id_vars=['Region name'], value_vars=[str(year) for year in range(1997, 2022)],
                          var_name='Year', value_name='GDHI')

# Convert 'Year' to integer
gdhi_long_chart['Year'] = pd.to_numeric(gdhi_long_chart['Year'])

# Plotting both GDHI and Population on separate subplots for clarity
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(16, 18), sharex=True)

# Plot GDHI Trends
sns.lineplot(data=gdhi_long_chart, x='Year', y='GDHI', hue='Region name', marker='o', ax=axes[0])
axes[0].set_title('GDHI Trends by Region (1997-2021)')
axes[0].set_ylabel('GDHI (in millions)')
axes[0].grid(True)

# Plot Population Trends
sns.lineplot(data=population_long_chart, x='Year', y='Population', hue='Region name', marker='o', ax=axes[1])
axes[1].set_title('Population Trends by Region (1997-2021)')
axes[1].set_xlabel('Year')
axes[1].set_ylabel('Population')
axes[1].grid(True)

# Adjust legend
axes[0].legend(title='Region name', bbox_to_anchor=(1.05, 1), loc='upper left')
axes[1].legend(title='Region name', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```

```python
# Calculate growth rates for GDHI and Population from the first year (1997) to the last year (2021) for each region
first_year = 1997
last_year = 2021

# Calculating initial and final values for GDHI and Population
initial_gdhi = total_GDHI.groupby('Region name')[str(first_year)].first()
final_gdhi = total_GDHI.groupby('Region name')[str(last_year)].first()
gdhi_growth = ((final_gdhi - initial_gdhi) / initial_gdhi) * 100

initial_population = total_res_pop.groupby('Region name')[str(first_year)].first()
final_population = total_res_pop.groupby('Region name')[str(last_year)].first()
population_growth = ((final_population - initial_population) / initial_population) * 100

# Combine growth rates into a single DataFrame for analysis
growth_data = pd.DataFrame({
    'GDHI Growth (%)': gdhi_growth,
    'Population Growth (%)': population_growth
}).reset_index()

# Plotting the growth rates using seaborn
plt.figure(figsize=(14, 10))
sns.scatterplot(data=growth_data, x='Population Growth (%)', y='GDHI Growth (%)', hue='Region name', s=100, palette='viridis')
plt.title('GDHI and Population Growth Rates by Region (1997-2021)')
plt.grid(True)
plt.axhline(0, color='gray', lw=1)
plt.axvline(0, color='gray', lw=1)
plt.xlabel('Population Growth (%)')
plt.ylabel('GDHI Growth (%)')
plt.xticks(rotation=45)
plt.legend(title='Region name', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

```

Here are the line charts for both Gross Disposable Household Income (GDHI) and population across all regions from 1997 to 2021:

GDHI Trends
- Overall Increase: Most regions, such as Wandsworth, Kensington and Chelsea, and Westminster, reflect general increases in GDHI, which indicates rising income levels across these affluent areas.
- Differential Growth: Regions like Hackney and Tower Hamlets have seen much steeper increases in GDHI. This may suggest significant economic development, possibly driven by urban regeneration projects and increased investment in these areas.

Population Trends
- Population Changes: Regions like Newham and Barking and Dagenham have experienced notable population growth, possibly due to housing developments and their relative affordability. Conversely, the City of London has seen a fluctuation and generally lower population growth, reflecting its nature as a business district with less residential housing.



Correlation with GDHI: The correlation between population growth and GDHI increase is evident in regions like Southwark and Greenwich, where both economic and demographic growth are aligned. This contrasts with areas like Richmond upon Thames, where population has remained relatively stable despite economic growth.

```python
# Calculate the Spearman's rank correlation matrix for GDHI and consumption categories
spearman_correlation_matrix = filtered_data.corr(method='spearman')

# Filter the correlation matrix to show only correlations with GDHI
spearman_gdhi_correlations = spearman_correlation_matrix["total_GDHI_2015 (in millions)"].sort_values(ascending=False)
spearman_gdhi_correlations
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the visual style
sns.set(style="white")

# Create a heatmap
plt.figure(figsize=(10, 8))
heatmap = sns.heatmap(spearman_correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=1)
heatmap.set_title('Spearman Correlation Heatmap between food Categories and income')
plt.show()
```

Positive Correlations:
- Tea and Coffee: The strongest positive correlation (0.25) with GDHI suggests that higher income areas might consume more tea and coffee.
- Water and Grains: These also show positive correlations, indicating that higher income could be associated with increased consumption of water and grain-based products.
- Fish and Poultry have positive correlations (0.19 and 0.15, respectively), suggesting that higher income levels might also be associated with higher consumption of these protein sources.

Negative Correlations:
- Dairy Products: There's a notably strong negative correlation (-0.26) with GDHI, indicating that higher income areas might consume less dairy.
- Wine and Fruit/Vegetables: These show moderate negative correlations, suggesting less consumption in higher income areas.

Higher income groups often have access to a broader variety of food and might prefer premium or health-oriented products such as organic grains or bottled water, which might explain the positive correlations with tea, coffee, water, and grains. Similarly, The stronger correlation of protein sources like fish and poultry with higher incomes could be tied to dietary preferences for higher-quality proteins or health-conscious eating habits prevalent among wealthier demographics.

The negative correlation with dairy could be related to dietary preferences or health consciousness in higher income groups, who may opt for alternative non-dairy options.
Similarly, the lower consumption of wine and fruits/vegetables in higher-income areas could reflect specific lifestyle choices or dietary trends prevalent among wealthier demographics.

The expectation might be that higher incomes correlate with increased consumption of all healthy or luxury food items, such as fruits, vegetables, and wine. However, the data shows a decrease in these areas, which could indicate regional preferences, accessibility issues, or a shift towards other types of luxury or specialty foods not captured in the typical categories.
